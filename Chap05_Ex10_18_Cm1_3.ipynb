```python
#5-3-1

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,num=11,endpoint=True)
y = np.cos(-x**2/5.0)

plt.plot(x,y,'o')
plt.grid(True)
```

```python
#5-3-1-1

from scipy import interpolate

f = interpolate.interp1d(x,y,'linear')
plt.plot(x,f(x),'-')
plt.grid(True)
```

```python
#5-3-1-2

f2 = interpolate.interp1d(x,y,'cubic')

xnew = np.linspace(0,10,num=30,endpoint=True)

plt.plot(x,y,'o', xnew,f(xnew),'-', xnew,f2(xnew),'--')

plt.legend(['data','linear','cubic'], loc='best')
plt.grid(True)
```

```python
#5-3-3-3

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def lorenz_func(v,t,p,r,b):
  return [-p*v[0] + p*v[1], -v[0]*v[2] + r*v[0] - v[1], v[0]*v[1] - b*v[2]]

p = 10
r = 28
b = 8/3
v0 = [0.1,0.1,0.1]
t = np.arange(0,100,0.01)

v = odeint(lorenz_func, v0, t, args=(p,r,b))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(v[:,0], v[:,1], v[:,2])

plt.title('Lorenz')
plt.grid(True)
plt.show()
```

```python
#5-3-4-1

from scipy.optimize import fsolve

def f(x):
  y = 2*x**2 + 2*x -10
  return y

x = np.linspace(-4,4)
plt.plot(x,f(x))
plt.plot(x,np.zeros(len(x)))
plt.grid(True)
```

```python
x1 = fsolve(f,2)
print(x1)

x2 = fsolve(f,-3)
print(x2)
```

```python
#5-3-4-2

from scipy.optimize import minimize

def objective(x):
  x1 = x[0]
  x2 = x[1]
  x3 = x[2]
  x4 = x[3]
  return x1*x4*(x1+x2+x3) + x3

def constraint1(x):
  return x[0]*x[1]*x[2]*x[3]-25.0

def constraint2(x):
  sum_sq = 40
  for i in range(4):
    sum_sq = sum_sq - x[i]**2
  return sum_sq

x0 = [1,5,5,1]
print(objective(x0))
```

```python
b = (1.0,5.0)
bnds = (b,b,b,b)
con1 = {'type':'ineq', 'fun': constraint1}
con2 = {'type':'ineq', 'fun': constraint2}
cons = [con1,con2]

sol = minimize(objective,x0,method='SLSQP',bounds=bnds,constraints=cons)
print(sol)
```

```python
print('Y:',sol.fun)
print('x:',sol.x)
```

```python
#総合問題5-3

def func(x):
  return x**2 + 1

def cons(x):
  return (x + 1)

cons = (
    {'type': 'ineq', 'fun': cons}
)

x = -10

result = minimize(func, x0=x, constraints=cons, method='SLSQP')
print(result)
```

```python
print('Y:',result.fun)
print('x:',result.x)
```
