```python
import numpy as np
import math

x = np.random.uniform(0.0, 1.0, 10000)
y = np.random.uniform(0.0, 1.0, 10000)
```

```python
in_x = []
in_y = []
out_x = []
out_y = []
count_in = 0

for i in range(0,10000):
  a = math.hypot(x[i],y[i])
  if a <1:
    count_in += 1

    in_x.append(x[i])
    in_y.append(y[i])

  else:
    out_x.append(x[i])
    out_y.append(y[i])

print("円の内部に入った数:",count_in)
```

```python
import matplotlib.pyplot as plt

circle_x = np.arange(0,1,0.001)
circle_y = np.sqrt(1- circle_x*circle_x)

plt.plot(circle_x,circle_y)

plt.scatter(in_x,in_y,color="b")
plt.scatter(out_x,out_y,color="y")

plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
```

```python
print("円周率", 4.0*count_in/10000)
```
